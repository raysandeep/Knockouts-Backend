```python
from json import JSONDecodeError
import requests
QUESTIONS_URL = "http://127.0.0.1:8000/api/admin/question/get/"
headers = {
    'authorization': 'Token 641c8d3b9fb9e9f615433e200948dec222d3df75',
    'content-type': 'application/json'
}
questions_resp = requests.request("GET", QUESTIONS_URL, headers=headers)
questions_resp = questions_resp.json()
```

```python
testcases = []
status=0
for i in questions_resp:
    TESTCASE_GET_URL = "https://api1.knockouts.dscvit.com/api/admin/testcase/get/" + i['id'] + "/"

    testcase_response = requests.request("GET", TESTCASE_GET_URL, headers=headers)
    try:
        testcase_response = testcase_response.json()
    except Exception as e:
        print(str(e),i['id'])
    if i['difficulty_level'] == 55:
        testcases.append({
        'id':i['id'],
        'difficulty_level':i['difficulty_level'],
        "question_title":i["question_title"],
        "question":i["question"],
        'testcases':testcase_response
    })
    status+=1
    print(status)
print(testcases[0])
```

```python
print(len(testcases))
TESTCASE_CREATE_URL = 'http://127.0.0.1:8000/api/admin/testcase/create/'
headers1 = {
    'authorization': 'Token 641c8d3b9fb9e9f615433e200948dec222d3df75',
    'content-type': 'application/json'
}
def uploadAllTestcases(question,obj):
    payload = {
        "question":question,
        "stdin":obj['stdin'],
        "stdout":obj['stdout'],
        "max_time":obj['max_time'],
        "max_memory":obj['max_memory'],
        "is_sample":obj['is_sample']
    }
    response = requests.request("POST", TESTCASE_CREATE_URL, headers=headers1, json=payload)
    print(response.status_code)
    # print(response.text)
```

```python
API_URL = 'http://127.0.0.1:8000/api/admin/question/create/'
status = 0
for i in testcases[0:5]:
    print(i["question_title"])
for i in testcases:
    payload = {
                'difficulty_level':i['difficulty_level'],
                "question_title":i["question_title"],
                "question":i["question"]
            }
    response = requests.request("POST", API_URL, headers=headers1, json=payload)
    print(response.status_code)
    resp = response.json()
    for j in i['testcases']:
        try:
            uploadAllTestcases(resp['id'],j)
        except Exception as e:
            print(str(e),resp['id'])
    status+=1
    print(status)
    print("---------")
```
